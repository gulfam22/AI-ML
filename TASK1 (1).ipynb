```python
plt.figure()
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.show()

plt.figure()
sns.scatterplot(data=df, x="sepal_width", y="petal_width", hue="species")
plt.title("Sepal Width vs Petal Width by Species")
plt.show()
```

```python
numeric_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], bins=15, kde=True)
    plt.title(f"Histogram: {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
    plt.figure()
sns.histplot(data=df, x="petal_length", hue="species", bins=15, kde=True, element="step", stat="count")
plt.title("Histogram: petal_length by Species")
plt.show()
plt.figure()
sns.boxplot(data=df[numeric_cols])
plt.title("Box Plots of All Numeric Features")
plt.show()
# Example: save one figure
plt.figure()
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.savefig("scatter_sepal_vs_petal.png", dpi=150, bbox_inches="tight")
plt.close()

```

```python
for col in numeric_cols:
    plt.figure()
    sns.boxplot(data=df, x="species", y=col)
    plt.title(f"Box Plot: {col} by Species")
    plt.xlabel("species")
    plt.ylabel(col)
    plt.show()
```
